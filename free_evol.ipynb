```python
from qutip import *
import matplotlib.pyplot as plt
import numpy as np
import math
from pylab import *
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
```

```python
#wektory stanu na sferze Blocha w kierunku x, y, z, -x, -y, -z
psi_z = basis(2, 0)
psi_mz = basis(2, 1)
psi_y = (basis(2, 0)+1j*basis(2, 1)).unit()
psi_my = (basis(2, 0)-1j*basis(2, 1)).unit()
psi_x = (basis(2, 0)+basis(2, 1)).unit()
psi_mx = (basis(2, 0)-basis(2, 1)).unit()
```

```python
Omega_x = 2*np.pi*0.05 # Parametr
H =  Omega_x * sigmax() # Hamiltonian
psi_z = basis(2, 0)
psi_mz = basis(2, 1)
psi0 = psi_mz #Psi(t=0) warunki początkowe
e_ops = [sigmax(), sigmay(), sigmaz(), psi_z.proj(), psi_mz.proj()] # wartosci oczekiwane beda liczone dla tych operatorow ktore są tu wyszczegolnione 
times = np.linspace(0.0, 30.0, 300)#
result = sesolve(H, psi0, times, e_ops, progress_bar=True)
sx=result.expect[0]
sy=result.expect[1]
sz=result.expect[2]
P_0=result.expect[3]
P_1=result.expect[4]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].plot(result.times, sx) 
axs[0].plot(result.times, sy)
axs[0].plot(result.times, sz)
axs[0].set_xlabel('Time') 
axs[0].set_ylabel('Expectation values') 
axs[0].legend(("$\sigma_X$", "$\sigma_Y$", "$\sigma_Z$"))
axs[0].grid()

axs[1].plot(result.times, P_0) 
axs[1].plot(result.times, P_1)
axs[1].set_xlabel('Time') 
axs[1].set_ylabel('Probabilities') 
axs[1].legend(("$P_{0}(t)$", "$P_{1}(t)$"))
axs[1].grid()
plt.show() 
```

```python
b = Bloch()
vec = [[1,0,0],[0,1,0],[0,0,1]]
pnts = [sx, sy, sz]
b.add_points(pnts)
b.add_vectors(vec)
b.show()
```

```python
#Uwzglednienie dekoherencji
gammax=0.15
gammay=0.15
gammaz=0.15
c_ops=gammax * sigmax()+gammay * sigmay()+gammaz * sigmaz()
```

```python
Omega_x = 2*np.pi*0.05 # Parametr
H =  Omega_x * sigmax() # Hamiltonian
psi_z = basis(2, 0)
psi_mz = basis(2, 1)
psi0 = basis(2, 0) #Psi(t=0) warunki początkowe
e_ops = [sigmax(), sigmay(), sigmaz(), psi_z.proj(), psi_mz.proj()] # wartosci oczekiwane beda liczone dla tych operatorow ktore są tu wyszczegolnione 
times = np.linspace(0.0, 30.0, 300)#
result = mesolve(H, psi0, times, c_ops, e_ops, progress_bar=True)
sx=result.expect[0]
sy=result.expect[1]
sz=result.expect[2]
P_0=result.expect[3]
P_1=result.expect[4]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].plot(result.times, sx) 
axs[0].plot(result.times, sy)
axs[0].plot(result.times, sz)
axs[0].set_xlabel('Time') 
axs[0].set_ylabel('Expectation values') 
axs[0].legend(("$\sigma_X$", "$\sigma_Y$", "$\sigma_Z$"))
axs[0].grid()

axs[1].plot(result.times, P_0) 
axs[1].plot(result.times, P_1)
axs[1].set_xlabel('Time') 
axs[1].set_ylabel('Probabilities') 
axs[1].legend(("$P_{0}(t)$", "$P_{1}(t)$"))
axs[1].grid()
plt.show() 
```
